```python
# Tree Traversal:
#-----------------

# Depth First Traversal
#------------------------
# 1. In-Order Traversal: Refers to visting the left nodes followed by root and then right nodes. (Left, Root, Right) 
#    Algorithm Inorder(tree)
   # a. Traverse the left subtree, i.e., call Inorder(left-subtree)
   # b. Visit the root.
   # c. Traverse the right subtree, i.e., call Inorder(right-subtree)

# 2. Pre-Order Traversal: refers to the visting root nodes followed by left nodes then right nodes. (Root, left, Right)
#    Algorithm Preorder(tree)
   # a. Visit the root.
   # b. Traverse the left subtree, i.e., call Preorder(left-subtree)
   # c. Traverse the right subtree, i.e., call Preorder(right-subtree)
    
# 3. Post-Order traversal: refers to visiting the left nodes followed by right nodes then root nodes. (Left, Right, Root)
#    Algorithm Postorder(tree)
   # a. Traverse the left subtree, i.e., call Postorder(left-subtree)
   # b. Traverse the right subtree, i.e., call Postorder(right-subtree)
   # c. Visit the root.
    
# Breadth First or Level Order Traversal : It will go sequencely level by level
```

```python
# Program for Tree traversal
#----------------------------

class Node: 
    def __init__(self,key): 
        self.left = None
        self.right = None
        self.val = key

# A function to do inorder tree traversal 
def Inorder(root): 
  
    if root: 
  
        # First recur on left child 
        Inorder(root.left) 
  
        # then print the data of node 
        print(root.val)
  
        # now recur on right child 
        Inorder(root.right) 


# A function to do postorder tree traversal 
def Postorder(root): 
  
    if root: 
  
        # First recur on left child 
        Postorder(root.left) 
  
        # the recur on right child 
        Postorder(root.right) 
  
        # now print the data of node 
        print(root.val) 
  
  
# A function to do preorder tree traversal 
def Preorder(root): 
  
    if root: 
  
        # First print the data of node 
        print(root.val)
  
        # Then recur on left child 
        Preorder(root.left) 
  
        # Finally recur on right child 
        Preorder(root.right)
        

# Driver code 
root = Node(1) 
root.left      = Node(2) 
root.right     = Node(3) 
root.left.left  = Node(4) 
root.left.right  = Node(5) 
print("Preorder traversal of binary tree :")
Preorder(root) 
  
print("\nInorder traversal of binary tree :")
Inorder(root) 
  
print("\nPostorder traversal of binary tree :")
Postorder(root)
```

```python
# SORTING :
#-------------

# 1. BUBBLE SORT : Refers to that works by repeatedly swapping the adjacent elements if they are in wrong order.

def bubbleSort(a):
    l = len(a)-1
    # Traverse through all the elements in the list
    for each in range(l):
        # Last each element already in place
        for x in range(l-each):
            # Traverse through all the elements from 0 to l-each 
            # Swap the elements if found element greater than next element
            if a[x] > a[x+1]:
                a[x], a[x+1] = a[x+1], a[x]
    return a

a = [85, 10, 182, 30]
sort_list = bubbleSort(a)
print("Find the sorting list by using Bubble Sort Algorithm :", sort_list)



# 2. INSERTION SORT : Refers to that works the way we sort playing cards in our hands.

def insertion_sort(nlist):

    # We start from 1 since the first element is trivially sorted
    for index in range(1, len(nlist)):
        currentValue = nlist[index]
        currentPosition = index

        # As long as we haven't reached the beginning and there is an element
        # in our sorted list larger than the one we're trying to insert - move
        # that element to the right
        while currentPosition > 0 and nlist[currentPosition - 1] > currentValue:
            nlist[currentPosition] = nlist[currentPosition -1]
            currentPosition = currentPosition - 1


        # We have either reached the beginning of the list or we have found
        # an element of the sorted list that is smaller than the element
        # we're trying to insert at index currentPosition - 1.
        # Either way - we insert the element at currentPosition
        nlist[currentPosition] = currentValue
    return nlist

nlist = [46,14,43,27,57,41,45,21,70]
insertion_list = insertion_sort(nlist)
print("\nFind the sorting list by using Insertion Sort Algorithm :", insertion_list)


# 3. SELECTION SORTING : The selection sort algorithm sorts an array by repeatedly finding the minimum element 
# (considering ascending order) from unsorted part and putting it at the beginning.

def selection_sort(alist):
    # create a loop with a loop variable i that counts from 0 to the length of the list – 1.
    for i in range(0, len(alist) - 1):
        smallest = i
        
        # an inner loop with a loop variable j that counts from i + 1 up to the length of the list – 1.
        for j in range(i + 1, len(alist)):
            # Inside the inner loop, if the elements at index j is smaller than the element at index smallest, 
            # then set smallest equal to j
            if alist[j] < alist[smallest]:
                smallest = j
        # After the inner loop finishes, swap the elements at indexes i and smallest
        alist[i], alist[smallest] = alist[smallest], alist[i]

    return alist
 
alist = [3,2,1,4,5,2,6]
selection_list = selection_sort(alist)
print("\nFind the sorting list by using Selection Sort Algorithm :", selection_list)
```

```python
# BINARY SERACH TREE
# --------------------
# Search a sorted array by repeatedly dividing the search interval in half. Begin with an interval covering the whole array.
# If the value of the search key is less than the item in the middle of the interval, narrow the interval to the lower half.
# Otherwise narrow it to the upper half. Repeatedly check until the value is found or the interval is empty.

def binarySearch(my_list, num):
    low = 0
    high = len(my_list)-1
    
    while low <= high:
        mid = (low + high) // 2
        
        # Check if num is present at mid 
        if my_list[mid] == num:
            return mid
        else:
            # If num is greater, ignore left half 
            if my_list[mid] < num:
                low = mid + 1
            else:
                # If num is smaller, ignore left half 
                high = mid -1
                
    # If we reach here, then the element was not present
    return False


my_list = [ 2, 3, 4, 10, 40 ] 
num = 10

# To call the function
result = binarySearch(my_list, num)

if result != False:
    print('Element %s is present at index for Binary Search % d' % (result, num))
else:
    print('Element %s is not present in list' % num)
    
    
# LINEAR SEARCH

def linearSearch(arr, x): 
  
    for i in range(len(arr)): 
  
        if arr[i] == x: 
            return i 
  
    return -1
        
arr = [10, 20, 80, 30, 60, 50, 110, 100, 130, 170]
x = 100

result = linearSearch(arr, x)

if result != False:
    print('\nElement %s is present at index for linear Search % d' % (result, x))
else:
    print('Element %s is not present in list' % x)
```

```python
# Quick sort:
# using pivot element as last element of the list
def partition(l, low, high):
    i = low - 1
    pivot = l[high]
    for j in range(low, high):
        if l[j] < pivot:
            i += 1
            l[i], l[j] = l[j], l[i]
    l[i+1], l[high] = l[high], l[i+1]
    return i+1


def quicksort(l, low, high):
    if low < high:
        pi = partition(l, low, high)

        quicksort(l, low, pi-1)
        quicksort(l, pi+1, high)


l = [10,8,9,2,1,5]
quicksort(l, 0, len(l)-1)
print('Quick Sort:',l)
```
