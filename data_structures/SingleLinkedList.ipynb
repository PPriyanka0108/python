```python
# Steps to add the nodes to the LinkedList

# 1. create the node
# 2. create the single linkedlist
# 3. add nodes to the linked list
# 4. print linked list


# create node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        # Initially our linkedlist is empty. So, head of the linked list is None
        self.head = None

    def insert(self, newNode):
        # Before inserting the newNode to the LinkedList, we need to see if the head of the LinkedList is empty.
        # If the LinkedList is empty then we have to make the newNode as Head of the LinkedList
        # headNode => John->None
        if self.head is None:
            self.head = newNode
        else:
            # headNode => John->Ben->None
            # here next of lastNode points to None i.e. lastNode=> ben->None
            # break the connection of next of last node and point the next of lastNode to the newNode
            lastNode = self.head
            while True:
                # Check if the next of lastNode is None the break
                if lastNode.next is None:
                    break
                # if the next of lastNode is not None i.e. headNode => John->Ben->None
                # lastNode is 'John' and lastNode.next is 'Ben' so, we are assigning lastNode.next to lastNode
                lastNode = lastNode.next
            # here the lastNode is 'Ben' and lastNode.next is now pointing to the newNode i.e. thirdNode which is (newNode)
            lastNode.next = newNode

    def printlist(self):
        # headNode => John->Ben->Eden->None
        # initially LinkedList points to the headNode which is 'John'
        if self.head is None:
            print('List is empty')
            return

        currentNode = self.head
        while True:
            if currentNode is None:
                break
            print(currentNode.data)

            # Now the currentNode points to the next node
            currentNode = currentNode.next



# Node => data, next
# data- is the value
# next - the pointer that points to the next node data

# firstNode.data => 'John' and firstNode.next => 'None'
# because first.Node is not pointing to any node as of now
# Create the firstNode
firstNode = Node('John')

# Initially our linkedlist is empty thats why we won't pass any parameter in the class LinkedList
linkedlist = LinkedList()

# Need to add the firstNode to the LinkedList
linkedlist.insert(firstNode)

# create the secondNode
secondNode = Node('Ben')
# Need to insert the secondNode to the LinkedList
linkedlist.insert(secondNode)

# create the thirdNode
thirdNode = Node('Eden')
# Need to insert the thirdNode to the LinkedList
linkedlist.insert(thirdNode)

# to print LinkedList
linkedlist.printlist()
```
